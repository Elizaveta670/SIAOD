## Стандартная функция сортировки




```python
def Sort_Po_Ymolch(array): 
    array.sort()
    return array
```

## Быстрая сортировка Хоара

```python
def quick_sort(array):
    #условие выхода 
    if len(array)<=1:
        return array
    #выбираем рандом.элемент
    elen = array[random.choice(range(0, len(array)))]
    #ген.списка и фильтр
    left = list(filter(lambda x: x<elen,array))
    #через ген.списка
    center= [i for i in array if i==elen]
    right = list(filter(lambda x: x>elen,array))
    
    return quick_sort(left) + center + quick_sort(right)
```

## Сортировка вставкой 

```python
def Vstavka(array):  
    # с 0 до последнего 
    for i in range(len(array)):  
        if i != len(array) - 1 and array[i] > array[i + 1]:
            #меняем местами
            array[i], array[i + 1] = array[i + 1], array[i]
            
            # с i до 0 шаг -1
            for j in range(i, 0, -1):  
                if array[j] <= array[j - 1]:
                    array[j - 1], array[j] = array[j], array[j - 1]
    return array
```

## Генерация массива

```python
import random
#ген. массива 
def random_array(length=100): 
    mas = []
    while len(mas) < length:
        mas.append(random.randint(0, 1000))
    return mas
#заполнение для проверки времени и вывод 
out_array = random_array(1_000) 
print(out_array)
```

## Оценка времени работы трех алгоритмов

```python
import timeit

# Сортировка массива и вывод времени
print("Сортировка по умолчанию   " + str(timeit.timeit("Sort_Po_Ymolch(out_array)", number=1, globals=globals())))
print("Быстрая сортировка        " + str(timeit.timeit("quick_sort(out_array)", number=1, globals=globals())))
print("Сортировка вставками      " + str(timeit.timeit("Vstavka(out_array)", number=1, globals=globals())))
```
